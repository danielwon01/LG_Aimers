Machine Learning - Data로 부터 내재된 패턴을 학습하는 과정 
지도학습은 정답레이블이 주어진 학습을 의미한다. 


지도학습  Machine Learning 문제
* Binary Classification  - 이산변수 출력
* Multiclass Classification - 이산변수 출력 
* Regression - 연속변수 출력





지도학습의 데이터셋 구성 
input x와 output y로 구성된다. 이러한 데이터를 통해 달성하고자 하는 목표는 바로 x -> y로 가는 함수 h를 학습, 즉 새로운 input이 들어왔을 때 함수 h를 맞추게 되는것을 의미 

지도학습에서는 실제 값과 예측 값의 차이인 에러를 줄여가면서 학습이 진행된다. 바로 학습 샘플에 레이블이 주어짐에 따라서 가능해지는 방법이다. 
테스트에서는 모델이 실제 환경에 적용되는 단계 이다. 이 때의 모델 정확도가 진정한 성능이 됨 

각 데이터에는 feature들이 존재하는데 이는 어느정도의 domain knowledge를 알고 있어야 용이하게 활용이 가능, 입력 feature의 선택은 machine learning 엔지니어가 주체적으로 할 수 있지만 관련 분야의 전문성을 가진 전문가와 함께 해결 가능하다. 딥러닝 같은 경우는 feature 자체적으로 학습함으로 무필요 




Hypothesis 함수의 학습 과정 
목적은 target Fuction F을 모사하는 것, 이러한 방법/수단으로서 real world에서 나타나는 샘플들의 realization으로서의 데이터셋을 활용한다. 
모델 학습과정에서는 크게 feature selection, model selection, optimization 과정을 거치게 된다. 

model selection 과정은 풀고자 하는 문제에 가장 적합한 model을 선택하는 과정 
optimization은 모델 파라미터를 최적화 하여 model이 가장 우수한 성능을 제공하도록 하는 것






머신러닝은 그 자체로 data의 결핍으로 인한 불확실성을 내포하고 있다. 학습과정에서 모든 데이터 샘플을 관찰할 수 없기 때문이다. 고로 머신러닝 모델 학습과정에서 중요한 점은 일반화 이다. 즉 모델이 학습과정에서 관찰하지 못한 샘플에 대해서도 우수한 성능을 제공할 수 있어야함 , 이러한 모델의 일반화된 성능을 측정하기 위한 측정방법으로 generalization error E을 정의, 이를 최소화 하는 것이 목적 , but 완전히 최소화하는 것이 어렵기 때문에 지도 학습에서는 트레이닝 에러, validation error, test error를 통해 generalization error를 최소화 하도록 노력 한다. 






Etrain은 모델을 주어진 데이터셋에 맞추어 학습하는데  사용하는 에러이다. 즉 주어진 샘플에서 모델 파라미터를 최적화 하는데 사용한다. 그렇게 때문에 이 에러는 E_general을 approxomation하는데 적합하지 않다. 대신에 train sample과 overlap이 되지 않도록 전체 데이터셋에서 일부 샘플을 따로 빼서 테스트 샘플을 정의한다. 이 샘플에서 나타나는 에러를 Etest라고 정의한다. 이 에러가 모델이 real world에서 적용될 때 나타나는 에러이다. 
목적은 Etest가 0으로 근사하게 만들어내는 것이다. 




모델의 정확도를 높이는 방법은 bias를 낮추고 모델 일반성을 높이고 variance를 낮추는 것이다. 각 task를 수행하기 위해서는 상반된 방법이 적용된다.   

bias 와 variance tradeoff 관계 존재 , 모델이 복잡해지면 overfitting 이 발생하기 쉬우며 높은 분산을 갖지만 낮은 편향을 나타낸다. 반대로 모델이 지나치게 단순하면 underfitting 발생하여 편향이 높아져 문제가 된다. 


딥러닝과 같은 오늘날의 머신러닝은 이미지나 비디오 등과 같은 고차원의 data를 다루기 때문에 복잡도가 굉장히 증가하고 있다. 하지만 복잡도가 증가하는 속도에 비해 데이터 샘플의 수가 쉽게 늘리지 못하는 문제가 존재하기 때문에 overfitting문제가 현저히 늘어나고 있다. 이는 입력 데이터 또는 입력 feature의 차원이 증가하면 지수적으로 샘플의 숫자가 늘어나야 하지만 그렇지 못하게 되며 **차원의 저주** 문제가 나타나고 있다.
